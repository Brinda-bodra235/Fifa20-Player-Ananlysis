### Why K-means ?

#### Fix those features having object datatype  ex. (83+2)

```python
# List of position columns to fix
position_cols = ['ls','st','rs','lw','lf','cf','rf','rw',
                 'lam','cam','ram','lm','lcm','cm','rcm','rm',
                 'lwb','ldm','cdm','rdm','rwb','lb','lcb','cb','rcb','rb']

# Function to fix these columns
def fix_position_values(data, cols):     # define a function for those cols
    for col in cols:
        if col in data.columns:
            data[col] = data[col].apply(             
                lambda x: sum(int(num) for num in str(x).split('+')) if '+' in str(x) else pd.to_numeric(x, errors='coerce') 
            )   
    return data

# apply lambda func and add thode values i.e (83+2 = 85), 85 will be output
```

```python
# Apply the function to replace original values
data = fix_position_values(data, position_cols)
```

```python
# Use for loop to handle outliers according to median 
for i in df:
    Q1 = np.quantile(df[i],0.25,method='midpoint')                   # first Quartile
    Q3 = np.quantile(df[i],0.75,method='midpoint')                   # third Quartile
    IQR = Q3 - Q1                                                    # Set Inter Quartile Range
    lower_limit = Q1 - (1.5 * IQR)                                       # Define Lower Limit/Bound
    upper_limit = Q3 + (1.5 * IQR)                                       # Define Upper Limit / Bound
    
    # Change Outlier's with median
    df.loc[(df[i]<lower_limit) | (df[i]>upper_limit),i] = df[i].median() 
```

## Principal Component Analysis :-
* It is used for high dimensionality data.
* It is used to reducing dimensionality oof the dataset.
* It is used to to create **new uncorrelated features (Principal Components)**, that maximize variance to capture essence of the data. 

```python
from sklearn.decomposition import PCA
pca = PCA()
Principal_components = pca.fit_transform(df)
Principal_components = pd.DataFrame(Principal_components)
```

```python
# Take first 3 PC 
PCA_DATA = Principal_components.iloc[:, :3]
PCA_DATA.head(3)
```

## Implementation of K-means: 
* **Clustering** : It is process of grouping the data based on similarity.

### How to decide the value of K? :
* 1. Domain Expert 
* 2. Data Visualisation
* 3. Elbow Technique

## Elbow Technique for Number of Clusters :-
* It is most used to determine the most optimal value of the K.
* Wherever we find the elbow or sharp angle, that will be the most optimal value of K.

### WCSS
* It is **Within Custer Sum of Square**
1. Square from datapoint to centroid 
2. Sum of all of them

```python
# find number of clusters 
from sklearn.cluster import KMeans     # Import the KMeans clustering algorithm from scikit-learn
WCSS = []                              # Initialize an empty list to store the Within-Cluster Sum of Squares (WCSS) values
for i in range(1,15):                  # Loop over different values of k (for 1 to 14)
    k_means = KMeans(n_clusters= i)    # Create a KMeans model with 'i' clusters
    k_means.fit(PCA_DATA)              # Fit the KMeans model on the PCA-transformed data
    WCSS.append(k_means.inertia_)      # Append the inertia_ (WCSS) or speed of calculation
```

```python
# Plot WCSS and no. of cluster map to find elbow point
plt.plot(range(1,15), WCSS, marker ="o")
plt.title("Elbow Technique")
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS")
```

* At 3 the elbow is noticed, where curve take the sharp curve
* Therefore, number of optimal cluster is 3

```python
# Cluster data for k = 3
from sklearn.cluster import KMeans
model = KMeans(n_clusters= 3, random_state= 99)
model.fit(PCA_DATA)
Labels = model.labels_   
Labels
```

* This will give which datapoint is in which cluster i.e 1,2 or 3
* For ex. First datapoint is in 2nd cluster. 

## Model Evaluation

```python
# Check Quality of Clusters
from sklearn.metrics import silhouette_score
Score_kmeans = silhouette_score(PCA_DATA, Labels)
Score_kmeans 
```

* It is **more than 0** so, the clusters are equally formed.
* To select the optimal number of clusters (e.g., in K-Means, try different **𝑘** and **pick the one with the highest average silhouette score**).

```python
# Perform k-means clustering with n_cluster = 3
k_means = KMeans(n_clusters=3)
k_means.fit(PCA_DATA)
labels = k_means.labels_      # coordinates of datapoints  
centroids = k_means.cluster_centers_   # coordinates of centroid 
```

```python
# Plot the Clusters 
plt.scatter(PCA_DATA.iloc[:, 0], PCA_DATA.iloc[:,1], PCA_DATA.iloc[:,2], c= labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='o', color = 'red')
plt.title("Clustering with n_clusters = 3")
plt.show()
```

## Conclusion 
1. The **small dots** are your data points after dimensionality reduction.

* **Yellow points** - belong to **Cluster 1**

* **Teal points** - belong to **Cluster 2**

2. The **red circles** represent the **cluster centroids** (average position of all points in that cluster).
